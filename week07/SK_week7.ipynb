```python
def solution(lottos, win_nums):
    win_nums, lottos = set(win_nums), set(lottos)
    print(lottos, win_nums)
    zeros = 7 - len(lottos) if 0 in lottos else 0
    correct = 6 - len(win_nums - lottos)
    best =  7 - correct - zeros if 7 - correct - zeros < 7 else 6
    worst = 7 - correct if 7 - correct < 7 else 6
    return [best, worst]
```

```python
def solution(enroll, referral, seller, amount):
    bill = {name:[i, 0] for i, name in enumerate(enroll)}
    for s, a in zip(seller, amount):
        ten = int(a*100*0.1)
        ninty = a*100 - ten
        bill[s][1] += ninty
        
        while True:
            head = referral[bill[s][0]]
            if head != '-':
                _ten = int(ten*0.1)
                if _ten < 1:
                    bill[head][1] += ten
                    break
                else:
                    bill[head][1] += ten - _ten
                    ten = _ten
            else:
                break
            s = head
            
    answer = [bill[i][1] for i in enroll]
    return answer
```
