```python
class Solution:
    def maximumUnits(self, boxTypes: List[List[int]], truckSize: int) -> int:
        boxTypes.sort(key=lambda x: x[1])
        ans = 0
        for _ in range(truckSize):
            if boxTypes[-1][0] == 0:
                boxTypes.pop()
            if len(boxTypes)==0:
                break
            boxTypes[-1][0] -= 1
            ans += boxTypes[-1][1]
        return ans
```

```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        mini = nums[0]
        maxi = nums[0]
        ans = nums[0]
        for i in range (1,len(nums)):
            x = max(nums[i],maxi*nums[i],mini*nums[i])
            y = min(nums[i],maxi*nums[i],mini*nums[i])
            maxi = x
            mini = y

            ans = max(ans, maxi)
        return ans
```
