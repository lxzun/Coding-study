<p>"ryan"이 "con"을 4번 신고했으나, 주어진 조건에 따라 한 유저가 같은 유저를 여러 번 신고한 경우는 신고 횟수 1회로 처리합니다. 따라서 "con"은 1회 신고당했습니다. 3번 이상 신고당한 이용자는 없으며, "con"과 "ryan"은 결과 메일을 받지 않습니다. 따라서 [0, 0]을 return 합니다.</p>

```python
from collections import defaultdict, Counter

def solution(id_list, report, k):
    id_list = {id:i for i, id in enumerate(id_list)}
    new_report = defaultdict(set)
    reported = defaultdict(set)
    
    for i in report:
        l = i.split()
        new_report[l[0]].update([l[1]])
        reported[l[1]].update([l[0]])
        
    reported = {i:len(reported[i]) for i in reported}
    blocked = set()
    for i in reported:
        if reported[i] >= k:
            blocked.update([i])
    
    answer = [0] * len(id_list)
    
    # print(reported, blocked, new_report)
    for i in new_report:
        names = set(new_report[i])
        answer[id_list[i]] += len(names & blocked)
    
    return answer
```

```python
case1 = [["muzi", "frodo", "apeach", "neo"], ["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"], 2, [2,1,1,0]]
case2 = [["con", "ryan"], ["ryan con", "ryan con", "ryan con", "ryan con"], 3, [0,0]]
for id_list, report, k, ans in [case1, case2]:
    print(f'id_list: {id_list}\n'
          f'report: {report}\n'
          f'k: {k}, ans: {ans}')
    print(f'Ans: {solution(id_list, report, k)==ans}\n')
```

```python
from collections import defaultdict
import math

def string2min(time_string):
    h, m = list(map(int, time_string.split(':')))
    return h*60+m

def min2string(min):
    h, m = min//60, min %60
    return '{:2d}:{:2d}'.format(h,m)

def solution(fees, records):
    new_records = defaultdict(list)
    for i in records:
        l = i.split()
        new_records[l[1]].append(string2min(l[0]))
    car_nums = list(new_records.keys())
    car_nums.sort()

    for key in new_records:
        total = 0
        if len(new_records[key]) %2 != 0:
            new_records[key].append(1439) # last time
            
        for idx in range(0,len(new_records[key]),2):
            total += new_records[key][idx+1] - new_records[key][idx]
                
        new_records[key] = total

    # print(new_records, car_nums)
    answer = []
    
    for num in car_nums:
        if new_records[num] <= fees[0]:
            cost = fees[1]
        else:
            cost = fees[1] + math.ceil((new_records[num] - fees[0])/fees[2]) * fees[3]
        answer.append(cost)
    return answer
```

```python
def draw_graph(info, edges):
    graph = {}
    for p,n in edges:
        if p not in graph:
            graph[p] = [[n], info[p]]
        else:
            if n not in graph[p][0]:
                graph[p][0].append(n)
    return graph


def DFS(graph, root, sw_count):
    visited = []
    stack = [root]
    before = []

    while stack:
        n = stack.pop()
        if n not in visited:
            if sw_count[0] > sw_count[1]-1:
                visited.append(n)
                
            sw_count[graph[n][1]] += 1
            graph[n][1] = None
            stack += graph[n] - set(visited)

            
            else:
                pass
            
    return visited


def solution(info, edges):
    sw_count  = {0:1, 1:0}
    graph = draw_graph(info, edges)
    answer = DFS(graph, 0, sw_count)
    print(graph)
    answer = 0
    return answer
```
