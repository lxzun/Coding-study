```python
import itertools
from collections import deque

def solution(priorities, location):
    answer = 0
    queue = deque()
    name = [_ for _ in range(len(priorities))]
    
    for name,i in zip(name,priorities):
        queue.append([name,i])

    while(queue):
        if len(queue) == 1:
            answer += 1
            break
        if max(j[1] for j in deque(itertools.islice(queue,1,len(queue)))) > queue[0][1]:
            target = queue.popleft()
            queue.append(target)
        else: # print
            target = queue.popleft()
            answer += 1
            if target[0] == location:
                break

    return answer
```
