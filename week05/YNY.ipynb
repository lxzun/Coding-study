```python
from string import ascii_lowercase

def solution(begin, target, words):
    answer = 0
    alpha_list = list(ascii_lowercase)
    while True:
        for idx,i in enumerate(begin):
            for j in alpha_list:
                tmp_word = list(begin)
                tmp_word[idx] = j
                tmp_word = "".join(tmp_word)
                if tmp_word in words:
                    answer += 1
                    begin = tmp_word

                    if begin == target:
                        return answer


                    else:
                        words.remove(begin)
                        break
            continue
        if len(words) == 0:
            return 0
```

```python
def solution(tickets):
    tickets.sort(key=lambda x: (x[0], x[1]))
    path = ['ICN']
    def dfs(tickets, path):
        if len(tickets) == 0: 
          return path
        cnt = 0
        while cnt < len(tickets) and tickets[cnt][0] != path[-1]:
            cnt += 1  # 시작 지점 idx 찾기
        if cnt == len(tickets):
            return []  # 끝까지 봤는데 없음

        while tickets[cnt][0] == path[-1]:
            answer = dfs(tickets[:cnt] + tickets[cnt + 1:], path + [tickets[cnt][1]])
            if answer != []:
                return answer
            cnt += 1
        return answer

    res = dfs(tickets, path)

    return res
```
