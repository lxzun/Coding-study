```python
from collections import defaultdict 
def solution(genres, plays):
    answer = []
    pair = defaultdict(list)
    for i,(j,k) in enumerate(zip(genres, plays)):
        pair[j].append((i,k))  # i:idx, j:장르, k:재생수
    
    num_plays = {}
    for n,m in pair.items():
        tmp_plays = 0
        for p in m:
            tmp_plays += p[1]
        num_plays[n] = tmp_plays
        
    num_plays = sorted(num_plays.items(), key=lambda x: -x[1])
    
    for a in num_plays:
        pair[a[0]].sort(key=lambda x: -x[1])
        
    

    for b in num_plays:
        class_name = b[0]
        cnt = 0
        for c in pair[class_name]:
            if cnt<2:
                answer.append(pair[class_name][cnt][0])
                cnt+=1
            else:
                continue
        
    return answer
```
