```python
class Solution:
    def finalPrices(self, prices: List[int]) -> List[int]:
        ans = []
        for i in range(len(prices)):
            flag = False
            for j in prices[i+1:]:
                if prices[i] >= j:
                    ans.append(prices[i]-j)
                    flag = True
                    break
                else:
                    pass
            if flag == False:
                ans.append(prices[i])
        return ans
```

```python
class Solution:
    def maxDistance(self, nums1: List[int], nums2: List[int]) -> int:
        ans = 0
        idx1,idx2=0,0
        while idx1 < len(nums1) and idx2 < len(nums2):
            if nums1[idx1]<=nums2[idx2]:
                ans = max(ans,idx2-idx1)
                idx2 +=1
                continue
            else:
                idx1+=1
                idx2+=1
        return ans
```
