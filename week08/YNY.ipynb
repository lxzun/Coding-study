```python
import numpy as np
def solution(rows, columns, queries):
    answer = []
    map = np.arange(1,rows*columns + 1).reshape(rows, columns)

    for query in queries:
        query = [x-1 for x in query]
        start = map[query[0],query[1]]  #시작점
        mini = start

        for i in range(query[0]+1, query[2]+1):            #왼쪽변
            map[i-1][query[1]] = map[i][query[1]]
            mini = int(min(mini,map[i][query[1]]))

        for i in range(query[1]+1, query[3]+1):             #아래변
            map[query[2]][i-1] = map[query[2]][i]
            mini = int(min(mini,map[query[2]][i]))

        for i in range(query[2]-1, query[0]-1, -1):         #오른쪽변
            map[i+1][query[3]] = map[i][query[3]]
            mini = int(min(mini,map[i][query[3]]))

        for i in range(query[3]-1, query[1]-1, -1):         #윗변
            map[query[0]][i+1] = map[query[0]][i]
            mini = int(min(mini,map[query[0]][i]))

        map[query[0]][query[1]+1] = start

        answer.append(mini)

    return answer
```
