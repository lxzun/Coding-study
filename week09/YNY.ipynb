```python
from itertools import combinations
from collections import Counter

def solution(orders, course):
    answer = []
    for idx in course:
        course_list = []
        for order in orders:
            tmp = combinations(sorted(order),idx)   # list(course 개수대로 만들수 있는 조합)
            course_list += tmp

        counter = Counter(course_list)
        if len(counter) != 0 and max(counter.values()) != 1:   # 이 조합 주문한사람이 없거나 혼자 주문한 경우 x
            answer += [''.join(c) for c in counter if counter[c] == max(counter.values())]  # 가장 많이 주문된 횟수의 메뉴들 출력
    return sorted(answer)
```

```python
from collections import defaultdict, deque

def solution(n, edge):
    graph= defaultdict(set)
    visit = [0]*(n+1)

    for i,j in edge:
        graph[i].add(j)
        graph[j].add(i)

    queue = deque([1])
    visit[1] = 1
    while queue:
        current = queue.popleft()
        for i in graph[current]:
            if visit[i] == 0:
                visit[i] = visit[current]+1
                queue.append(i)

    return visit.count(max(visit))
```
